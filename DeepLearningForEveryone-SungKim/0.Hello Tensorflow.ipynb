**Tensorflow사용**  
큰 단계는
* 그래프 구조 만들기 (computaional graph)
* 데이터를 그래프 안에 feed 하고 그래프를 run
* 그래프 안의 variables 업데이트

**그래프 구조 만들기**  
그래프는 **노드**로 구성  
노드를 생성 후 다음 노드에서 이전 노드의 정보를 사용하는 식으로  
복잡하게 짜여진 구조 생성  

*constant: 상수(변하지 않는 것)*

*tensorflow 1.0v 에서는 session을 사용하였으나 tensorflow 2.x 부터는 session없이 memory에 직접적으로 저장함*

```python
# tf.constant 사용 예시 1
import tensorflow as tf
node1 = tf.constant(3.0)
node2 = tf.constant(4.0)
node3 = tf.add(node1,node2)
```

```python
print(node3)
```

```python
#tf.constant 사용예시 2
hello = tf.constant("Hello World")
```

```python
print(hello)
```
